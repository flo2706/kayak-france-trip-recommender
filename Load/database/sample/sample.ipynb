```python
import pandas as pd
import os

# Input CSV file path
INPUT_FILE = os.path.join("..", "data", "hotels_weather_final_ter.csv")

# Output CSV file path
OUTPUT_FILE = "hotels_weather_final_ter_sample.csv"

# Read the original CSV file
df = pd.read_csv(INPUT_FILE)

# Get the first 5 rows as a sample
df_sample = df.head(5)

# Save the sample to a new CSV file
df_sample.to_csv(OUTPUT_FILE, index=False, encoding="utf-8")

print(f"Sample created: {OUTPUT_FILE}")
```
